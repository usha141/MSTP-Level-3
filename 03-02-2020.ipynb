### Calculate Probability of Playing
- P(No/Overcast)=(P(Overcast)==>Eq(1)
 1. Calculate Prior Probabilities
      - P(Overcast)==4/14==0.29
 2. Calculate Posterior Probabilities
      - P(Yes)=9/14=0.64
 3. Put Prior and Postrior probabilities in Eq(1)
      - P(Yes/Overcast)=0.44*0.64/0.29=0.98(Higher)
 

### Probability of not playing
- P(No/Overcast)=P(Overcast/No)P(No)/P(Overcast)===>Eq(2)
   1. Calculate Prior Probabilities:
       - P(Overcast)=4/14=0.29
       - P(No)=5/14=0.36
   2. Calculate Posterior Probabilities:
       - P(Overcast/No)=0/9=0
   3. Put Prior and Posterior probabilities in Eq(2)
       - P(No/Overcast)=0*0.36/0.29=0

### Probability of Playing
  - P(Play=yes/Overcast,Mild)=P(Overcast,Mild/yes)*P(No)===>Eq(1)
  - P(Overcast,Mild/Play=No)=P(Overcast/No)*P(Mild/No)===>Eq(2)
 1. calculate prior probabilities
       - p(Play=yes)=9/14=0.64
 2. calculate posterior probabilities
       - p(overcast/play=yes)=4/9=0.44 P(Mild/Play=yes)=4/9=0.44
 3. Put Posterior probabilities in Eq(2)
       - P(Overcast,Temp=Mild/yes)=0.44*0.44=0.1936(higher)
 4. Put prior and posterior probabilities in Eq(1)
       - P(Play=Yes/Overcast,Temp=Mild)=0.64*0.1936=0.124=0

### Probability of not Playing
  - P(Play=No/Overcast,Mild)=P(Overcast,Mild/No)*P(No)===>Eq(3)
  P(Overcast,Mild/Play=No)=P(Overcast/No)*P(Mild/No)===>Eq(4)
 1. calculate prior probabilities
       - p(Play=no)=5/14=0.36
 2. calculate posterior probabilities
       - p(overcast/play=No)=0/9=0 P(Mild/Play=No)=2/5=0.4
 3. Put Posterior probabilities in Eq(4)
       - P(Overcast,Temp=Mild/No)=0*0.4=0
 4. Put prior and posterior probabilities in Eq(3)
       - P(Play=No/Overcast,Temp=Mild)=0*0.36=0

### Encoding Feature
 - You need to convert these string labels into numbers
 ex: 'Overcast','Rainy','Sunny'as 0,1,2
 - Scikit-Learn provides "LabelEncoder"library for encoding labels between 0 and one less than the no of discrete classes

```python
data=pd.read_csv("Tennis.csv")
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
data['Outlook']=le.fit_transform(data['Outlook'])
data['Temperature']=le.fit_transform(data['Temperature'])
data['Humidity']=le.fit_transform(data['Humidity'])
data['Wind']=le.fit_transform(data['Wind'])
data['Play Tennis']=le.fit_transform(data['Play Tennis'])
```

```python
data
```

```python
X=data[['Outlook','Temperature','Humidity','Wind']]
Y=data[['Play Tennis']]
```

### Apply Navie Bayes Model

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gNB = GaussianNB()
gNB.fit(X,Y)
y_predict=gNB.predict(X)
```

```python
y_predict
```

```python
gNB.predict([[0,0,1,0]])
```

### Check Accuracy

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(Y,y_predict)
```
