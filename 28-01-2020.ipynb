```python
X=df['Head Size']#input
Y=df['Brain Weight']#output
```

### Line equation Formula
- Y=mX+c
- Y=c+mX
- Y=b0+b1*X
    - c or b0 is Intercept
    - m or b1 is a slope

## Slope Formula(b1)
- b1=((X-X_mean)*(Y-Y_mean))/(X-X_mean)^2

```python
n=len(X)
n
```

```python
n=len(X)
X_mean=np.mean(X)
print(X_mean)
Y_mean=np.mean(Y)
Y_mean
```

```python
numerator = 0
denaminator = 0
for i in range(n):
    numerator=numerator + (X[i] - X_mean)*(Y[i]-Y_mean)
    denaminator=denaminator+(X[i]-X_mean)**2
b1=numerator/denaminator
print(b1)#slope
```

## Intercept Formula(b0)
- b0=Y_mean-(b1*X_mean)


```python
b0=Y_mean-(b1*X_mean)
print(b0)
```

```python
max_X=np.max(X)
min_X=np.min(X)
max_X,min_X
```

```python
x=np.linspace(min_X,max_X,1000)
y=b0+b1*x
x
```

```python
y
```

```python
plt.scatter(X,Y,color='y',label='points')
plt.plot(x,y,color='g',label='line')
plt.title('Linear \nRegression')
plt.legend()
plt.show()
```

## R2 Calculation(Regression sum of square)
- r2=1-((Y-y_bar)**2/(Y-y_mean)**2)

```python
numerator2=0
denaminator2=0
for i in range(n):
    y_prediction=b0+b1*X[i]
    numerator2=numerator2+(Y[i]-y_prediction)**2
    denaminator2=(Y[i]-Y_mean)**2
r2=1-(numerator2/denaminator2)
```

```python
print(y_prediction)
print(r2)
```
